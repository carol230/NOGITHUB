```python
x = np.linspace(0, 2 * np.pi, 100)
f = np.sin(x)
def bisection(f, a, b, tol=1e-5):
    if f(a) * f(b) >= 0:
        raise ValueError()
    
    c = a
    while (b - a) / 2.0 > tol:
        c = (a + b) / 2.0
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c

root_bisection = bisection(lambda x: np.sin(x), 1,-1)
print(root_bisection)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3


xs = [-1, 1]
ys = [1, 3]
s = [
    {"a": 1, "b": -1, "c": 0.5, "d": 0.25},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.show()


```
